### 高次偏導関数 <span style="background-color:#bbc8e6 ; color: white ;">&nbsp;2.4[2]&nbsp;</span> <span style="background-color:#e79db0 ; color: black ;">&nbsp;140, 141&nbsp;</span> ###

関数 $f(x,y)$ 偏導関数もまた2変数関数となるので, それらを偏微分することができる.

#### 例）$f(x,y)=\sin(xy)$ ####

偏導関数 $f_x, f_y$ はそれぞれ

```
JupyterImageResolution = 72;
D[Sin[x*y],x]
```

```
D[Sin[x*y],y]
```

$f_x$ を$x, y$ でそれぞれ偏微分した偏導関数を $f_{xx}, f_{xy}$ と書く.
偏微分を計算すると

```
D[D[Sin[x*y],x],x]
```

```
D[D[Sin[x*y],x],y]
```

一方, $f_y$ を$x, y$ でそれぞれ偏微分した偏導関数を $f_{yx}, f_{yy}$ と書く.
偏微分を計算すると

```
D[D[Sin[x*y],y],x]
```

```
D[D[Sin[x*y],y],y]
```

これらを $f(x,y)$ の<b>2次偏導関数</b>という.
この例では, $f_{xy}=f_{yx}$ であることに気づくだろう.
実際に多くの場合, $f_{xy}=f_{yx}$ が成り立つ.
厳密には, 次のことが成り立つ.

<b>定理.</b>　$f_{xy}$ と $f_{yx}$ が点 $(a,b)$ でともに連続ならば, $f_{xy}(a,b)=f_{yx}(a,b)$ が成り立つ.

よって, 2次偏導関数は $f_{xx}, f_{xy}\,(=f_{yx}), f_{yy}$ の3つの関数である.
では, 3次偏導関数はいくつあるだろうか？ $n$次偏導関数は？

### 合成関数 $F(t) = f(x(t), y(t))$ の微分 <span style="background-color:#bbc8e6 ; color: white ;">&nbsp;2.4[3] 〜p.60&nbsp;</span> <span style="background-color:#e79db0 ; color: black ;">&nbsp;146, 147&nbsp;</span> ###

 - 1変数関数の微積では, 合成関数の微分の公式を学んだ;
 $$\frac{d}{dx}\left\{f(g(x))\right\}=f'(g(x))\cdot g'(x)$$

- この公式により, 一見複雑な関数も単純な関数の合成とみることによって計算が容易になることがわかる. つまり, 合成関数の微分の公式は「微分計算のテクニック」の一つであった.
- 2変数関数においても, 合成関数が定義でき, （偏）微分の公式が知られているが, 計算のテクニックとしての側面は弱く, むしろ, その<b>合成関数がもつ意味を理解すること</b>が重要である.

#### 2変数関数 $f(x,y)$ と, 2つの1変数関数 $x=\varphi(t), y=\psi(t)$ の合成関数 ####

$F(t):=f(\varphi(t), \psi(t))$ は, 独立変数 $t$ の1変数関数であるため, $t$ に関する導関数を考えることができる.
$$
F'(t)=f_x(\varphi(t), \psi(t))\,\varphi'(t)+f_y(\varphi(t), \psi(t))\,\psi'(t)
$$

$$
\frac{dF}{dt}=\frac{\partial f}{\partial x}\,\frac{dx}{dt}+\frac{\partial f}{\partial y}\,\frac{dy}{dt}
$$

- 2つの１変数関数の組 $(\varphi(t), \psi(t))$ は, $t$ に対して平面の点 $(\varphi(t), \psi(t))$ を対応させる関数とみることができる.
- $t$ の値を変化させるとき, 対応する点の全体は平面内の曲線をなす.
- $(\varphi(t), \psi(t))$ をこの曲線の<b>媒介変数表示</b>, または<b>パラメータ表示</b>という.

$(x,y)=(a+ht, b+kt)$ で与えられる曲線は, <b>点 $(a,b)$ を通り, ベクトル $(h,k)$ に平行な直線</b>である.

```
ParametricPlot[{1-t, 2+t/2},{t,-2,2}]
```

$(x,y)=(r\,\cos t, r\,\sin t)$ で与えられる曲線は, <b>原点 $(0,0)$ を中心とする半径$r$ の円周</b>である.

```
ParametricPlot[{2*Cos[t], 2*Sin[t]},{t,0,2*Pi}]
```

合成関数 $F(t):=f(\varphi(t), \psi(t))$ は, 関数 $f(x,y)$ を平面内の曲線 $(x,y)=(\varphi(t), \psi(t))$ に制限することと解釈できる.

```
Show[
    Plot3D[Sin[x*y]+3,{x,-3,3},{y,-3,3},PlotStyle->Opacity[0.3],Mesh->False],
    ParametricPlot3D[{1-t, 2+t/2,0},{t,-2,2}],
]
```

```
Show[
    Plot3D[Sin[x*y]+3,{x,-3,3},{y,-3,3},PlotStyle->Opacity[0.3],Mesh->False],
    ParametricPlot3D[{1-t, 2+t/2,0},{t,-2,2}],
    ParametricPlot3D[{1-t, 2+t/2,Sin[(1-t)*(2+t/2)]+3},{t,-2,2}]
]
```

```
Plot[Sin[(1-t)*(2+t/2)]+3,{t,-2,2}]
```

#### 2変数関数 $f(x,y)$ と, 2つの2変数関数 $x=\varphi(u, v), y=\psi(u, v)$ の合成関数 ####

$F(u,v):=f(\varphi(u,v), \psi(u,v))$ は, 独立変数 $u,v$ の2変数関数であるため, $u,v$ に関する偏導関数を考えることができる.
\begin{align*}
F_u(u,v)=&f_x(\varphi(u,v), \psi(u,v))\,\varphi_u(u,v)+f_y(\varphi(u,v), \psi(u,v))\,\psi_u(u,v)\\
F_v(u,v)=&f_x(\varphi(u,v), \psi(u,v))\,\varphi_v(u,v)+f_y(\varphi(u,v), \psi(u,v))\,\psi_v(u,v)
\end{align*}

\begin{align*}
\frac{\partial F}{\partial u}=&\frac{\partial f}{\partial x}\,\frac{\partial x}{\partial u}+\frac{\partial f}{\partial y}\,\frac{\partial y}{\partial u}\\
\frac{\partial F}{\partial v}=&\frac{\partial f}{\partial x}\,\frac{\partial x}{\partial v}+\frac{\partial f}{\partial y}\,\frac{\partial y}{\partial v}
\end{align*}

- 対応 $(x,y)=(\varphi(u,v), \psi(u,v))$ は, $uv$-座標系と$xy$-座標系の関係式と解釈することができる. つまり, これは座標の変換である.
- つまり,　合成関数 $F(u,v):=f(\varphi(u,v), \psi(u,v))$ を考えることは, 関数 $f(x,y)$ を $uv$-座標系の関数としてみることと解釈できる.
