### 2変数関数のそのグラフ <span style="background-color:#bbc8e6 ; color: white ;">&nbsp;2.3[1]&nbsp;</span> <span style="background-color:#e79db0 ; color: black ;">&nbsp;134, 135, 136&nbsp;</span> ###

<b>関数 $f(x,y)$ のグラフ</b>とは, 「方程式 $z=f(x,y)$ を満たす空間内の点 $(x,y,z)$ の全体」である.

#### 例）$f(x,y)=\sin(xy)$ ####

$x=-1, y=-\pi/4$ のとき, $z=f(-1,-\pi/4)=1/\sqrt{2}$ である.

```
Sin[(-1)*(-Pi/4)]
```

よって, 点$(-1, -\pi/4, 1/\sqrt{2})$ は, $z=\sin(xy)$ のグラフ上の点である.

```
a=-1;
b=-Pi/4;
Show[
    Plot3D[Sin[x*y],{x,-2,2},{y,-2,2},PlotStyle->Opacity[0.05],Mesh->False],
    Graphics3D[{PointSize[0.01],Point[{a,b,Sin[a*b]}],Point[{a,b,0}]}],
    Graphics3D[{Dashed, Line[{{a,0,0},{a,b,0},{0,b,0}}],Line[{{a,b,0},{a,b,Sin[a*b]}}]}],
    Graphics3D[{Arrow[{{-2,0,0},{2,0,0}}],Arrow[{{0,-2,0},{0,2,0}}],Arrow[{{0,0,-2},{0,0,2}}]}]
]
```

$z=\sin(xy)$ のグラフ上の点を他にも描いてみる.

```
Show[
    Graphics3D[Table[Point[{a,b,Sin[a*b]}],{a,-2,2,1/2},{b,-2,2,1/2}]]
]
```

もう少し

```
Show[
    Graphics3D[Table[Point[{a,b,Sin[a*b]}],{a,-2,2,1/5},{b,-2,2,1/5}]]
]
```

もう少し…

```
Show[
    Graphics3D[Table[Point[{a,b,Sin[a*b]}],{a,-2,2,1/10},{b,-2,2,1/10}]]
]
```

このような点の全体は, 空間内の<b>曲面</b>をなす.

```
Plot3D[Sin[x*y],{x,-2,2},{y,-2,2},PlotRange->All]
```

### 2変数関数の微分 <span style="background-color:#bbc8e6 ; color: white ;">&nbsp;2.4[1]&nbsp;</span> <span style="background-color:#e79db0 ; color: black ;">&nbsp;137, 138, 139&nbsp;</span> ###

- 2変数関数 $f(x,y)$ に対し, 一方の変数を定数とみて, 他方の変数について微分することを「<b>偏微分する</b>」という.
- 偏微分した関数のことを<b>偏導関数</b>とう.
- 『偏』は「かたより」ということ. 英語では "Partial derivative" または "Partial differentiation" という（「部分的な」微分）.
- $\dfrac{\partial}{\partial x}$ は「$x$ の関数とみて偏微分せよ」という記号と解釈する. $\dfrac{d}{dx}$ との違いを理解すること.
- 偏導関数を表す記号は「$\partial$」を使うものの他に, 微分する変数を関数記号の右下に書く流儀もある.

```
D[Sin[x*y],x]
```

```
D[Sin[x*y],y]
```
